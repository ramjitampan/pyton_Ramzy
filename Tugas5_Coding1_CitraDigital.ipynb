```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pywt

# ============================
# Baca gambar grayscale
# ============================
img = cv2.imread('/content/FAHMI.png', 0)

# ============================
# Contoh 1: Fourier Transform (FFT)
# ============================
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20 * np.log(np.abs(fshift))

plt.figure(figsize=(10,4))
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Spektrum Magnitudo (FFT)')
plt.show()

# ============================
# Contoh 2: Discrete Cosine Transform (DCT)
# ============================
img_float = np.float32(img) / 255.0
dct = cv2.dct(img_float)
idct = cv2.idct(dct)

plt.figure(figsize=(12,4))
plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title("Asli")
plt.subplot(132), plt.imshow(np.log(abs(dct)), cmap='gray'), plt.title("DCT")
plt.subplot(133), plt.imshow(idct, cmap='gray'), plt.title("Rekonstruksi")
plt.show()

# ============================
# Contoh 3: Wavelet Transform
# ============================
coeffs2 = pywt.dwt2(img, 'haar')
LL, (LH, HL, HH) = coeffs2

plt.figure(figsize=(10,8))
plt.subplot(221), plt.imshow(LL, cmap='gray'), plt.title('Approximation (LL)')
plt.subplot(222), plt.imshow(LH, cmap='gray'), plt.title('Horizontal Detail (LH)')
plt.subplot(223), plt.imshow(HL, cmap='gray'), plt.title('Vertical Detail (HL)')
plt.subplot(224), plt.imshow(HH, cmap='gray'), plt.title('Diagonal Detail (HH)')
plt.show()
```
