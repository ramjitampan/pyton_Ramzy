# Catatan kegunaan (skiprows=4)
> skiprows=4 supaya langsung baca data yang bener-bener tabel, bukan tulisan tambahan dari file World Bank. Setelah itu, datanya bisa di olah,karena biasanya file dari World Bank ada bagian header/keterangan yang bukan data (kayak nama sumber, catatan, dll).


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

%matplotlib inline
df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv', skiprows=4)
pd.set_option('display.max_columns', None)  # Tampilkan semua kolom
df.head(10)
```

```python
df.set_index('Country Name', inplace=True)
df.head()
```

```python
countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan']
df.TOP5 = df.loc[countries]
df.TOP5
```

```python
years = list(map(str, range(1960, 2020)))
print(years)
```

```python
df_top5 = df.TOP5[years].transpose()
df_top5.head()
```

```python
plt.figure(figsize=(14, 8))
plt.stackplot(df_top5.index, df_top5['China'], df_top5['India'], df_top5['United States'], df_top5['Indonesia'], df_top5['Pakistan'],
              labels=['China', 'India', 'United States', 'Indonesia', 'Pakistan'])

plt.title('Top 5 World Population', size=18)
plt.xlabel('\nYears', size=14)
plt.ylabel('Population', size=14)
plt.xticks(rotation=90)
plt.legend(fontsize=14, loc='upper left')
plt.margins(0.0)
plt.show()
```

```python
df_top5.plot(kind='area', figsize=(14, 8))

plt.title('Top 5 World Population', size=18)
plt.xlabel('\nYears', size=14)
plt.ylabel('Population', size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(fontsize=14)
plt.show()
```

```python
colors = ['magenta', 'purple', 'steelblue', 'green', 'lime']

df_top5.plot(kind='area', figsize=(14, 8), color =colors)

plt.title('Top 5 World Population', size=18)
plt.xlabel('\nYears', size=14)
plt.ylabel('Population', size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(fontsize=14)
plt.show()
```

```python
df_top5.plot(kind='area', figsize=(14, 8), cmap='plasma')

plt.title('Top 5 World Population', size=18)
plt.xlabel('\nYears', size=14)
plt.ylabel('Population', size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(fontsize=14)
plt.show()
```

```python
df_top5.plot(kind='area', figsize=(14, 8), cmap='plasma', alpha=0.5)

plt.title('Top 5 World Population', size=18)
plt.xlabel('\nYears', size=14)
plt.ylabel('Population', size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(fontsize=14)
plt.show()
```

```python
df_top5.plot(kind='area', figsize=(14, 8), cmap='plasma', alpha=0.5, stacked=False)

plt.title('Top 5 World Population', size=18)
plt.xlabel('\nYears', size=14)
plt.ylabel('Population', size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(14,8))

df_full = df_top5.divide(df_top5.sum(axis=1), axis=0)

plt.stackplot(df_full.index,
    [df_full['China'], df_full['India'],
    df_full['United States'], df_full['Indonesia'], df_full['Pakistan']],
    labels=['China', 'India', 'United States', 'Indonesia', 'Pakistan'],
    alpha=0.8)

plt.title('Top 5 World Population', size=18)
plt.xlabel('\nYear', size=14)
plt.ylabel('Total Population\n', size=14)
plt.xticks(rotation=90)
plt.yticks(size=14)
plt.legend(fontsize=14, loc='upper right')
plt.margins(0,0)
plt.show()
```

```python
df_full_top5 = df_top5.divide(df_top5.sum(axis=1), axis=0)

df_full_top5.plot(kind='area', figsize=(14,8), cmap='plasma', alpha=0.5)

plt.title('Top 5 World Population', size=18)
plt.xlabel('\nYear', size=14)
plt.ylabel('Total Population\n', size=14)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(fontsize=14, loc='lower left')
plt.show()
```
