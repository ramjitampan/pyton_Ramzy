```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca citra
img = cv2.imread("/content/RAMZY FOTO.jpg")
h, w, c = img.shape
print("Resolusi:", w, "x", h)
print("Kedalaman Warna (channel):", c)

# 2. Konversi Model Warna
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)       # RGB
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)       # HSV
img_ycbcr = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)   # YCbCr

# Visualisasi hasil konversi
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(img_rgb)
ax[0].set_title("RGB")
ax[0].axis("off")

ax[1].imshow(img_hsv[:, :, 0], cmap='hsv')
ax[1].set_title("HSV (Hue)")
ax[1].axis("off")

ax[2].imshow(img_ycbcr[:, :, 0], cmap='gray')
ax[2].set_title("YCbCr (Y)")
ax[2].axis("off")

plt.show()

# 3. Sampling & Kuantisasi
# Sampling (perkecil ukuran gambar jadi 64x64)
small = cv2.resize(img, (64, 64), interpolation=cv2.INTER_LINEAR)

# Kuantisasi warna (mengurangi level warna)
quantized = (img // 64) * 64

# Simpan hasil
cv2.imwrite("sampling.jpg", small)
cv2.imwrite("quantized.jpg", quantized)

# Tampilkan hasil
fig2, ax2 = plt.subplots(1, 2, figsize=(10, 5))
ax2[0].imshow(cv2.cvtColor(small, cv2.COLOR_BGR2RGB))
ax2[0].set_title("Sampling (64x64)")
ax2[0].axis("off")

ax2[1].imshow(cv2.cvtColor(quantized, cv2.COLOR_BGR2RGB))
ax2[1].set_title("Kuantisasi Warna")
ax2[1].axis("off")

plt.show()
```
